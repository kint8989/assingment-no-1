```python
1. Write a program to accept percentage from the user and display the grade according to the following 
criteria: 
Marks 
Grade 

>90 
A

>80 and <=90 
B 
>=60 and <=80 
C 
below 60 
D 
ANS 
percentage = float(input("Enter percentage: "))
if percentage > 90:
    print("Grade A")
elif percentage > 80 and percentage <= 90:
    print("Grade B")
elif percentage >= 60 and percentage <= 80:
    print("Grade C")
else:
    print("Grade D")
    
2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the 
following criteria: 
Tax 
15% 
10% 
5% 
Cost Price(in Rs) 
>100000 
5 
50000 and <= 100000 
<= 50000 

ANS=
cost_price = float(input("Enter cost price of the bike (in Rs): "))
if cost_price > 100000:
    road_tax = cost_price * 0.15
elif cost_price <= 100000 and cost_price > 50000:
    road_tax = cost_price * 0.10
else:
    road_tax = cost_price * 0.05
print("Road tax to be paid is:", road_tax)

3. Accept any city from the user and display monuments of that city. 
City 
Delhi 
Monument 
Red Fort 
Taj Mahal 
Agra 
Jaipur 
Jal Mahal 

ANS=
city = input("Enter city name: ")
if city == "Delhi":
    print("Monuments: Red Fort, Taj Mahal")
elif city == "Agra":
    print("Monument: Taj Mahal")
elif city == "Jaipur":
    print("Monument: Jal Mahal")
else:
    print("No monuments found for the given city")
    
4. Check how many times a given number can be divided by 3 before it is less than or equal to 10. 

num = int(input("Enter a number: "))
count = 0
while num > 10:
    num = num // 3
    count += 1
print("The number can be divided by 3 for", count, "times before it is less than or equal to 10")


5. Why and When to Use while Loop in Python give a detailed description with example 
The while loop is used in Python when a set of instructions or a block of code needs to be repeated until a certain condition is met. The loop iterates through the block of code repeatedly until the condition becomes false. A while loop can be used when we do not know the number of iterations beforehand, unlike a for loop.
For example, let's say we want to take input from the user until they enter a valid number between 1 and 10. We can use a while loop to repeat the input prompt until the user enters a valid number:
num = 0
while num < 1 or num > 10:
    num = int(input("Enter a number between 1 and 10: "))
print("You entered", num)

6. Use nested while loop to print 3 different pattern.
# Pattern 1
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1

# Pattern 2
i = 5
while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i -= 1

# Pattern 3
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print(j, end="")
        j += 1
    print()
    i += 1

7. Reverse a while loop to display numbers from 10 to 1. 
ANS=num = 10
while num >= 1:
    print(num)
    num -= 1
    
    
8. Reverse a while loop to display numbers from 10 to 1


ANS=for num in range













```
